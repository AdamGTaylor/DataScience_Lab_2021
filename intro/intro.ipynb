```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import pathlib
import h5py

#path
#path1 = pathlib.Path().absolute()
#print(str(path1))

#toc
from nbindex import floating
floating.tableofcontent()
floating.codehider()

def read_hdf5(path):

    weights = {}

    keys = []
    with h5py.File(path, 'r') as f: # open file
        f.visit(keys.append) # append all keys to list
        for key in keys:
            if ':' in key: # contains data if ':' in key
                print(f[key].name)
                weights[f[key].name] = f[key].value
    return weights
```

## Source

<font size="4">The data is coming from the International Space Station Lightning Imaging Sensor (ISS LIS). A near real time data is avaliable. The sensor takes images from S:54° to N:54° in lattitude and from -180° to 180 in longitude of Earth's surface. The data format used to store the processed images are .h5 which is a hierarchical dat format. The algorithm that processes the taken pictures gone under multiple enhancements and the data is considered validated.
