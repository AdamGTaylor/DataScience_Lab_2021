```python
import pandas as pd
import numpy as np

import sys
import os
import pathlib
import glob

import netCDF4

from sklearn import metrics

sys.path.append("src/")

from visualization import mpsettings
import matplotlib.pyplot as plt
import shap

from catboost import CatBoostRegressor

import matplotlib

plt.style.use("classic")

matplotlib.rcParams["figure.figsize"] = (8,8)
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams["axes.labelsize"] = 20
matplotlib.rcParams["axes.grid"] = True
matplotlib.rc('lines',linewidth = 2)
```

```python
def my_first_plot():
    plt.figure(figsize=(16,10))
    plt.scatter(Eddy["radius"].values,Eddy["Longevity"].values, alpha = .4,color ="red")
    plt.xlabel("Calculated relative radius")
    plt.ylabel("Longevity")
    plt.title(f"Eddy longevity-radius correlation\n From the Aviso {experiment} observation")

    plt.show()
    
def my_second_plot():
    #plt.figure(figsize=(16,10))
    plt.plot(y_test, model.predict(x_test), " bo", alpha=.7)
    plt.plot([-100,1300], [-100,1300], ":r")
    plt.xlabel("Original Longevity")
    plt.ylabel("Predicted Longevity")
    plt.title("Observed and predicted Longevity\n R$^2$ score: {0:.3f}".format(metrics.r2_score(y_test,model.predict(x_test))))

    plt.xscale("log")
    plt.yscale("log")

    plt.xlim(1,1200)
    plt.ylim(1,600)

    plt.show()
    
def my_third_plot():
    plt.figure(figsize=(16,10))
    plt.plot(range(len(x_test)),pd.Series(model.predict(x_test)).sort_values(), label = "predicted")
    plt.plot(range(len(x_test)),y_test.sort_values("Longevity"), label = "observed", alpha = .7)
    plt.xlabel("Identifier")
    plt.ylabel("Longevity")
    plt.title("Observed and predicted Longevity")
    plt.legend(loc="best")
    
    plt.show()
```

```python
PATH = "data/"
fpath ="data"
experiment = "2005-2012"

Eddy = pd.read_csv(
    PATH +
    fpath+".csv"
    )
```

# Mesoscale Ocean Eddies along the U.S. West Coast

This project is about finding and following the different Eddy currents through multiple days and see if there a correlation between their longevity and radius.

## The Longevity - radius correlation

```python
my_first_plot()
```

```python
model = CatBoostRegressor(25000,l2_leaf_reg=7)

train = Eddy.sample(int(len(Eddy)*0.9))
test = Eddy.drop(train.index)

x_train = pd.read_csv(PATH +"x_train.csv")
y_train = pd.read_csv(PATH +"y_train.csv")

x_test = pd.read_csv(PATH +"x_test.csv")
y_test = pd.read_csv(PATH +"y_test.csv")


model.fit(x_train,y_train,verbose=False,plot=True)
```

```python
shap.initjs()
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(x_train)

id1 = 78
id2 = 102
id3 = 420
```

## Shap's figures

And how much I started to like them

```python
shap.force_plot(explainer.expected_value, shap_values[id1,:], x_train.iloc[id1,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values[id2,:], x_train.iloc[id2,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values[id3,:], x_train.iloc[id3,:])
```

```python
shap.summary_plot(shap_values,x_train,show=False)

fig = plt.gcf()
fig.set_figheight(10)
fig.set_figwidth(16)

ax = plt.gca()
```

```python
shap.dependence_plot("radius",shap_values,x_train,show=False)

fig = plt.gcf()
fig.set_figheight(10)
fig.set_figwidth(16)

ax = plt.gca()
```
